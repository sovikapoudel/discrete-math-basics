In this part, I am going to show the closest country from 'Nepal' where the single node means it's a neighbour country.

And for this I am going to use Graphiviz library which is a open source module for python to create graphs consisting of nodes and edges.

```python
dot = graphviz.Graph(comment='null') #.Graph gives undirected graph
dot.node('N', 'Nepal')
dot.node('I', 'India')
dot.node('C', 'China')
dot.node('P', 'Pakistan')

dot.edges(['NI','NC', 'IP']) #creating the link/edges between neighbour countries 
dot.edge('I', 'C', constraint='false')

dot
```
